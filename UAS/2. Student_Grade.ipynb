```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Menampilkan semua data awal
data = pd.read_csv('C:\\Users\\Ezi\\Tugas-Dasar-Pemrograman\\UAS\\student_grade.csv')
data
```

```python
#Menampilkan semua data dengan menambah kolom baru
data['description'] = ''
data.loc[data['grade'] >= 70,'description'] = 'PASS'
data.loc[data['grade'] <= 70,'description'] = 'FAIL'

data
```

```python
#Menampilkan kolom student number, name dan mean(rata-rata)
data_group= data.groupby(['student number','name']).mean().astype(int)
data_group = data_group[['grade']].rename(columns={'grade':'mean'})
data_group
```

```python
#menampilkan visualisasi

data_group.plot(kind='bar')
plt.xlabel('Name')
plt.ylabel('Grade')
plt.title('Mean of Final Exam Values')

plt.show()
```
